```python
#Batting average function
def batting_average(df, playerID):
    batting_average = df.loc[playerID, "H"] / df.loc[playerID, "AB"]
    return batting_average
```

```python
barrys_bat_avg = batting_average(batting_df, barrys_id)
```

```python
type(barrys_bat_avg)
```

```python
# Slugging percentage function
def slugging_percentage(df, playerID):
    slugging_percentage = (df.loc[playerID, "H"] + df.loc[playerID, "2B"] 
        + 2 * df.loc[playerID, "3B"] + 3 * df.loc[playerID, "HR"]) / df.loc[playerID, "AB"]
    return slugging_percentage
```

```python
barrys_slug_pct = slugging_percentage(batting_df, barrys_id)
```

**Step #10, Plot Barry Bonds Career** (15pts)

- Make a simple plot that shows home runs versus Barry Bonds age. Be sure that it has a well title, y-axis, x-axis.

- Be mindful of the x-limits and y-limits, make sure it looks nice

- Can you insert a vertical line that indicates the time when Barry Bonds is suspected of starting to use PEDs?

```python
fig, ax = plt.subplots()

# line graph of number of home runs vs age
ax.plot(barrys_stats.loc[:1993].age, barrys_stats.loc[:1993].home_runs, 
        c = "k", label = "Pirates", linewidth = 3)
ax.plot(barrys_stats.loc[1993:].age, barrys_stats.loc[1993:].home_runs, 
        c = "darkorange", label = "Giants", linewidth = 3)

# title and axis labels
ax.set_title("Barry Bonds: Home Runs vs Age", fontsize = 14, fontweight = "bold")
ax.set_xlabel("Age")
ax.set_ylabel("Home Runs")

# setting axis limits
ax.set_xlim(20, 45)
ax.set_ylim(barrys_stats.home_runs.min()-1, barrys_stats.home_runs.max()+1)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

#Inserting vertical line
age_ped = barrys_stats.loc[1998, "age"]
ax.axvline(x = age_ped,               # vertical line at age when started using PEDs
           color = "b",               # make the color blue
           label = "Speculated start of PED use",   # label (shows up in the legend)
           linestyle = ":",           # the line style
           linewidth = 2)             # thickness of the line

#Inserting horizontal line
avg_hr = barrys_stats.loc[:1997, "home_runs"].mean()
ax.axhline(y = avg_hr,                # horizonal line at average home runs
           color = "r",
           linestyle = "--",
           linewidth = 2)

#Legend
ax.legend(frameon = False)

# Adding text to describe above numbers
message = "Avg HR before PED use: \n" + str(round(avg_hr,2))
                                                  
ax.text(45,                           # placement on the x-axis
        28,                           # placement on the y-axis
        message,                      # text to show value
        horizontalalignment = 'left')
```

**Step #11, Scale it up** (10pts)

- Create a function that will take a variable name and generate a plot consistent with your work above and output the figure to a `.png` file.

- Describe where you are writing the .png files to, i.e. what folder are the showing up in?
    - the figure is saved to the same folder that this notebook is in
    
- Do this for home runs (like above), batting average, slugging percentage. 

- Note these three graphs need not have all the embellishments, it is most important that the graphs be (1) consistently done and (2) informative.

---

```python
def scaled_plotting_func(var):
    
    fig, ax = plt.subplots()

    # line graph of var (y) vs age (x)
    ax.plot(barrys_stats.loc[:1993].age, barrys_stats[var].loc[:1993], 
            c = "k", label = "Pirates", linewidth = 3)
    ax.plot(barrys_stats.loc[1993:].age, barrys_stats[var].loc[1993:], 
            c = "darkorange", label = "Giants", linewidth = 3)

    # title and axis labels
    if (var == "bat_avg"):
        print_var = "Bat Average"
    elif (var == "slug_pct"):
        print_var = "Slugging Percent"
    elif (var == "home_runs"):
        print_var = "Home Runs"
    
    ax.set_title("Barry Bonds: " + print_var + " vs Age", fontsize = 14, fontweight = "bold")
    ax.set_xlabel("Age")
    ax.set_ylabel(print_var)

    #Setting axis limits
    ax.set_xlim(20, 45)
    ax.set_ylim(0, barrys_stats[var].max()+0.2)

    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)

    #Inserting vertical line
    age_ped = barrys_stats.loc[1998, "age"]
    ax.axvline(x = age_ped,               # vertical line at age when suspected to have started
               color = "b",               # make the color blue
               label = "Suspected Start of PED use",   # label (shows up in the legend)
               linestyle = ":",           # the line style
               linewidth = 2)             # thickness of the line
    
    #Inserting horizontal lines
    avg_var = barrys_stats[var].loc[:1997].mean()
    ax.axhline(y = avg_var,               # horizonal line at average var
               color = "r",
               alpha = 0.75,
               linestyle = "--",
               linewidth = 2)

    #Legend
    ax.legend(frameon = False, loc = "upper left")
    
    # text to describe above numbers
    message = "Average " + print_var + "\nbefore PED use: " + str(round(avg_var,2))

    ax.text(45,                           # placement on the x-axis
            0.95*avg_var,                 # placement on the y-axis
            message,                      # text to show value
            horizontalalignment = 'left')
    
    # output the figure to .png file
    plt.savefig("barrys_stats_" + var + ".png", bbox_inches = "tight", dpi = 600)
```

```python
scaled_plotting_func("bat_avg")
```

```python
scaled_plotting_func("slug_pct")
```

```python
scaled_plotting_func("home_runs")
```
